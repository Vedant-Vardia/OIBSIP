### Importing Libraries & Nessasary Functions

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt
import calendar 
```

#### Removing Warnings

```python
import warnings
warnings.filterwarnings('ignore')
```

#### Importing Data

```python
df = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
df
```

```python
df.describe()
```

#### Checking For Null Values.

```python
df.isnull().sum()
```

#### Renaming 'Region' to 'State' and 'Region.1' to 'Region'

```python
df=df.rename(columns={df.columns[0]:"State",df.columns[6]:"Region"})
df.head()
```

```python
df['State'].unique()
```

#### Checking No of States & Data on each state

```python
df.groupby("State").size()
```

#### Checking The no of Regions

```python
df.groupby("Region").size()
```

#### Using date to find the Month

```python
df['Date']=pd.to_datetime(df[' Date'],dayfirst=True)
df['Month']=df['Date'].dt.month
df['Month_int']=df["Month"].apply(lambda x:int(x))
df['Month']=df['Month_int'].apply(lambda x:calendar.month_abbr[x])
df.head()
```

```python
df.groupby("Month").size()
```

## Estimated Employment Rate 

```python
fig=px.bar(df,x='Region',y=' Estimated Labour Participation Rate (%)',color='State'
           ,title="Monthly Employment(Region-Wise)",animation_frame="Month",template='plotly')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

## Unemployment rate (Region-Wise)

```python
fig=px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',color='State',
           title="Monthly Unemployment Rate ",animation_frame='Month',template='plotly')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

# Unemployment Rate (State-Wise)

```python
fig = px.box(df,x='State',y=' Estimated Unemployment Rate (%)',
             color='State',title='Estimated Unemployment rate',
             template='plotly')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
fig=px.bar(df,x='State',y=' Estimated Unemployment Rate (%)',color='Region'
           ,title="Monthly Unemployment Rate(State-Wise)",animation_frame="Month",template='plotly')
fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```

## Complete Unemployment Rate Graph

```python
x=df[['State','Region',' Estimated Unemployment Rate (%)']]
fig=px.sunburst(x,path=['Region','State'],
               values=' Estimated Unemployment Rate (%)',
               width=800,height=800,color_continuous_scale="YlGnBl",
               title="Unemployment Rate In India")
fig.show()
```
